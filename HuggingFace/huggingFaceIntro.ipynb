```python
!pip install transformers
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")
classifier("Deadpool movie is really fun to watch")
```

```python
classifier(["Deadpool movie is really fun to watch",
            "i really like concept of multiverse in Flash movie",
            "green lantern movie was not upto the mark"])
```

```python
classifier_zeroShot = pipeline("zero-shot-classification")
classifier_zeroShot("I am planning to visit goa, for holidays",
                    candidate_labels=["education", "trip","politics"])
```

```python
generator = pipeline("text-generation")
generator("In the animation course, they will teach you how to")
```

```python
generator_distilgpt2 = pipeline("text-generation",model="distilgpt2")
generator_distilgpt2("In the animation course, they will teach you how to",
                     max_length=30,
                     num_return_sequences=3)
```

```python
unmasker=pipeline("fill-mask")
unmasker("DC comics is <mask> than other comics",
         top_k=2)
```

```python
ner=pipeline("ner", grouped_entities=True)
ner("My name is Clark Kent and I work in Daily Planet in New York")
```

```python
question_answerer=pipeline("question-answering")
question_answerer(
    question="were does Clar Kent works?",
    context="My name is Clark Kent and I work in Daily Planet in New York"
)
```
